```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Use forward slashes in the path
image = Image.open("C:/Users/SNEHALATHA/OneDrive/Desktop/traffic.jpg")

# Display the image
image.show()


# Convert the image to a numpy array
image = np.array(image)

# Check the dtype of the image
print("Original dtype:", image.dtype)

# If the dtype is not uint8 or float32, convert it
if image.dtype != np.uint8 and image.dtype != np.float32:
    image = image.astype(np.float32)

# If the dtype is float32 and values are greater than 1, normalize it to [0, 1]
if image.dtype == np.float32 and np.max(image) > 1.0:
    image = image / 255.0

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Turn off axes for a cleaner view
plt.show()

plt.rcParams["figure.figsize"] = (40,10)
image=cv2.imread("airport.jpg")
plt.imshow(image)
```

Import Libraries and Use pixellib built in function to create object and load model

```python
import pixellib
from pixellib.instance import instance_segmentation
segment_image = instance_segmentation()
segment_image.load_model("mask_rcnn_coco.h5")
```

Use segmentImage function for image segmentation, pass input image and output image name, and make show_bboxes parameter True for bounding box object detection

## Image Segementation and Object Detection

```python
segment_image.segmentImage("airport.jpg", output_image_name = "image_new.jpeg",show_bboxes = True)
image=cv2.imread("image_new.jpeg")
plt.rcParams["figure.figsize"] = (50,10)
plt.imshow(image)
```
